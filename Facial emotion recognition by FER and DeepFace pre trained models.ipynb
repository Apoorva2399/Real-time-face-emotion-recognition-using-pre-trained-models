```python
def rectangle(img) :
 gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
 # print (facecascade.empty())
 faces= faceCascade.detectMultiScale(gray_image,1.1,4)

 # draw rectangle on the face
 for (x,y,w,h) in faces :
  cv2.rectangle(image, (x,y),(x+w, y+h), (0, 255, 0), 2)
  Text =cv2.FONT_HERSHEY_SIMPLEX
  # use putText() to insert Text on the video
  res=cv2.putText(image, predictions['dominant_emotion'],(0,150),Text,1,(255,0,0), 2, cv2.LINE_4) ;
 return res
```

```python
# Real time face emotion recognition
```

```python
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

cap = cv2.VideoCapture(1)
#check webcam is opened correctly
if not cap.isOpened():
  cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("cannot open webcam")

while True:
    ret, frame = cap.read()

    result = DeepFace.analyze(frame, actions=['emotion'])

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    #print( faceCascade.empty() )
    faces = faceCascade.detectMultiScale(gray,1.3,5)

    #draw rectangle around face  
    for(x,y,w,h) in faces:
        cv2.rectangle(frame, (x,y),(x+w,y+h),(0,255,0), 2)
  
    font = cv2.FONT_HERSHEY_SIMPLEX
    #use putText() to insert text on video
    cv2.putText(frame,result['dominant_emotion'], (48,48), font,3, (0,0,255),2, cv2.LINE_4)
    cv2.imshow('Original video',frame)
   
    if cv2.waitKey(2) & 0xFF == ord('q') :
        break
    
cap.release()
cv2.destroyAllWindows()
```
